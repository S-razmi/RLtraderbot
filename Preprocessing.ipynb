```python
from datetime import datetime, timedelta

def datetime_range(start, end, delta):
    current = start
    while current < end:
        yield current
        current += delta

dts = [dt.strftime('%Y-%m-%d') for dt in 
       datetime_range(datetime(2022, 11,6), datetime(2023, 2, 15), 
       timedelta(days=1))]

# print(dts)
```

```python
import sys
import numpy as np
import csv
import pandas as pd
import regex as re
from tqdm import tqdm
import requests
import os
from datetime import datetime as dt
import pytz
directory = '/content/ETHUSDT'
rows = []
save_path = os.path.join(directory,'concat_file')
if not os.path.exists(save_path):
  os.mkdir(save_path)
for i in tqdm(dts):
  file_path = os.path.join(directory,f'ETHUSDT-1m-{i}.csv')
  # try:
  with open(file_path, 'r') as file:
      csvreader = csv.reader(file,delimiter=',')
      header = next(csvreader)
      for row in csvreader:
        rows.append(row)

df = pd.DataFrame(rows)
file_name = 'ETHUSDT.csv'
df.to_csv(os.path.join(save_path,file_name))
print(f'file {os.path.join(save_path,file_name)} created successfully')
```

```python
columns = ['Open time','Open','High','Low','Close','Volume','Close Time','Quote Asset Volume','Number of Trades','Taker buy base asset volume','Taker buy quote asset volume','Ignore']
df.columns = columns
df['Open time'] = df['Open time'].astype('int').apply(lambda x:dt.fromtimestamp(x/1000,tz=pytz.UTC))
df['Close Time'] = df['Close Time'].astype('int').apply(lambda x:dt.fromtimestamp(x/1000,tz=pytz.UTC))
df.drop(labels= ['Close Time','Ignore'],axis=1,inplace= True)
df.to_csv('/content/drive/MyDrive/RL Final Project/ETHUSDT.csv')
df.head()
```
